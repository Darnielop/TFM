```python
import pandas as pd
```

```python
symptoms = pd.read_csv('.\datasets\Diseases_Symptoms_Processed.csv')

treatments = pd.read_csv('.\datasets\Diseases_Treatments_Processed.csv')
```

```python
symptoms.columns = [col[3:] if col.startswith('or ') else col for col in symptoms.columns]
```

```python
for i in range(0, len(symptoms.columns)):
    print(symptoms.columns[i])
```

```python
duplicate_columns = symptoms.columns[symptoms.columns.duplicated()].tolist()
if duplicate_columns:
    print(f"Columnas duplicadas: {duplicate_columns}")
else:
    print("No hay columnas duplicadas.")
```

```python
# Identificar las columnas duplicadas
duplicate_columns = symptoms.columns[symptoms.columns.duplicated()].tolist()

# Eliminar las columnas duplicadas, manteniendo solo la primera aparicin
symptoms = symptoms.loc[:, ~symptoms.columns.duplicated()]

# Mostrar las columnas restantes
print(symptoms.columns)
```

```python
duplicate_columns = symptoms.columns[symptoms.columns.duplicated()].tolist()
if duplicate_columns:
    print(f"Columnas duplicadas: {duplicate_columns}")
else:
    print("No hay columnas duplicadas.")
```

```python
for i in range(0, len(symptoms.columns)):
    print(symptoms.columns[i])
```

```python
# Obtener las tres primeras columnas
fixed_columns = symptoms.columns[:3].tolist()

# Obtener el resto de las columnas y ordenarlas alfabticamente
sorted_columns = sorted(symptoms.columns[3:])

# Combinar las columnas fijas con las columnas ordenadas
new_column_order = fixed_columns + sorted_columns

# Reordenar el DataFrame con el nuevo orden de columnas
symptoms = symptoms[new_column_order]

# Mostrar las columnas reordenadas
print(symptoms.columns)
```
